# Análise da vetorização das notícias

## Imports, definição de funções e carregamento dos dados

```python

import matplotlib.pyplot as plt
from b_limpar.limpador import Limpador
from b_vetorizar.vetorizador import Vetorizador
from _common import util

noticias_df = util.carregar_todas_as_noticias() 
Limpador().limpar_se_necessario(noticias_df)
vetorizador = Vetorizador()
```

## Vetorização com o Google Universal Sentence Encoder

### Utilizando o texto da notícia _in natura_

#### Sem incluir o resumo da notícia

```python
vetorizador.vetorizar_use(noticias_df, usar_texto_limpo=False, incluir_resumo=False)
conteudo_reduzido = vetorizador.reduzir_dimensionalidade(noticias_df)
plt.scatter(conteudo_reduzido[:,0], conteudo_reduzido[:,1])
plt.show()
```

#### Incluindo o resumo da notícia

```python
vetorizador.vetorizar_use(noticias_df, usar_texto_limpo=False, incluir_resumo=True)
conteudo_reduzido = vetorizador.reduzir_dimensionalidade(noticias_df)
plt.scatter(conteudo_reduzido[:,0], conteudo_reduzido[:,1])
plt.show()
```

### Utilizando o texto da notícia limpo

#### Sem incluir o resumo da notícia

```python
vetorizador.vetorizar_use(noticias_df, usar_texto_limpo=True, incluir_resumo=False)
conteudo_reduzido = vetorizador.reduzir_dimensionalidade(noticias_df)
plt.scatter(conteudo_reduzido[:,0], conteudo_reduzido[:,1])
plt.show()
```

#### Incluindo o resumo da notícia

```python
vetorizador.vetorizar_use(noticias_df, usar_texto_limpo=True, incluir_resumo=True)
conteudo_reduzido = vetorizador.reduzir_dimensionalidade(noticias_df)
plt.scatter(conteudo_reduzido[:,0], conteudo_reduzido[:,1])
plt.show()
```
